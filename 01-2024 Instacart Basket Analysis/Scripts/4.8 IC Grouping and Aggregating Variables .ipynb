# Spending flag 

```python
#creating max_order column -> grouping user_id by prices and returning mean
df_ords_prods_merge['avg_spend'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#checking output
df_ords_prods_merge.head(100)
```

```python
#spending flag 
df_ords_prods_merge.loc[df_ords_prods_merge['avg_spend'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
#spending flag 
df_ords_prods_merge.loc[df_ords_prods_merge['avg_spend'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
# value counts 
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
#checking data 
df_ords_prods_merge[['user_id', 'spending_flag', 'avg_spend']].head(60)
```

# Frequency flag 

```python
#creating med_days_since_last_order column -> grouping user_id by days_from_last_order and returning median 
df_ords_prods_merge['med_days_since_last_order'] = df_ords_prods_merge.groupby(['user_id'])['days_from_last_order'].transform(np.median)
```

```python
#Frequency flag 
df_ords_prods_merge.loc[df_ords_prods_merge['med_days_since_last_order'] > 20, 'Frequency_flag'] = 'Non-Frequent Customer'
```

```python
#Frequency flag
df_ords_prods_merge.loc[(df_ords_prods_merge['med_days_since_last_order'] <= 20) & (df_ords_prods_merge['med_days_since_last_order'] > 10), 'Frequency_flag'] = 'Reglar customer'
```

```python
#Frequency flag 
df_ords_prods_merge.loc[df_ords_prods_merge['med_days_since_last_order'] <= 10, 'Frequency_flag'] = 'Frequent customer'
```

```python
# value counts 
df_ords_prods_merge['Frequency_flag'].value_counts(dropna = False)
```

```python
#checking data 
df_ords_prods_merge[['user_id', 'Frequency_flag', 'med_days_since_last_order']].head(60)
```

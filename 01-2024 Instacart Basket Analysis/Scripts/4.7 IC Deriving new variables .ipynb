# Defining the function

```python
#defining the function 
def price_label(row):
    if row['prices'] <=5:
        return 'Low-range product'
    elif (row['prices'] >5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

# applying the function 

```python
#applying price_label function 
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
#checking value counts
df['price_range'].value_counts()
```

```python
#checkingmax 
df['prices'].max()
```

# For-loops 

```python
#checking frequency orders_day_of_week to find busiest_day
df_ords_prods['orders_day_of_week'].value_counts(dropna = False)
```

```python
#creating the forloop
result = []

for value in df_ords_prods['orders_day_of_week']:
    if value == 0: 
        result.append('Busiest day')
    elif value == 4: 
        result.append('Least busy')
    else: 
        result.append('Regularly busy') 
```

```python
result
```

```python
#creating a column called busiest day that = result of forloop
df_ords_prods['busiest_day'] = result 
```

```python
#checking frequency
df_ords_prods['busiest_day'].value_counts(dropna = False)
```

# Two Busiest days column 

```python
#checking frequency orders_day_of_week to find busiest_day
df_ords_prods['orders_day_of_week'].value_counts(dropna = False)
```

```python
#busiest days = saturday and sunday and slowest days = wednesday and thursday
```

```python
#creating the forloop for 2 busiest days 
two_busiest_days = []

for value in df_ords_prods['orders_day_of_week']:
    if value == 0 or value == 1: 
        two_busiest_days.append('Busiest day')
    elif value == 4 or value == 3: 
        two_busiest_days.append('Least busy')
    else: 
        two_busiest_days.append('Regularly busy') 
```

```python
#creating a column called two_busiest_days that = result of forloop
df_ords_prods['two_busiest_days'] = two_busiest_days 
```

# Busiest hour of the day 

```python
#checking frequency orders_hour_of_day to find busiest hours
df_ords_prods['order_hour_of_day'].value_counts(dropna = False)
```

```python
#forloop for busiest hours the day 
busiest_hour = []

for value in df_ords_prods['order_hour_of_day']:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        busiest_hour.append('Most Orders')
    elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
        busiest_hour.append('Average orders')
    else: 
        busiest_hour.append('Fewest orders')
```

```python
#checking output 
busiest_hour
```

```python
#creating a column called busiest_period_of_day that = result of forloop
df_ords_prods['busiest_period_of_day'] = busiest_hour 
```

# Step 3 

creating a new column 'region' based on state column to determine differences in spending habits 

```python
#creating new column region using loc function with mapping 

df_ords_prods['region'] = 'West'

df_ords_prods.loc[df_ords_prods['state'].isin(['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'Pennsylvania', 'New Jersey']), 'region'] = 'Northeast'

df_ords_prods.loc[df_ords_prods['state'].isin(['Wisconsin', 'Michigan', 'Illinois', 'Indiana', 'Ohio', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri']), 'region'] = 'Midwest'

df_ords_prods.loc[df_ords_prods['state'].isin(['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma', 'Texas', 'Arkansas', 'Louisiana']), 'region'] = 'South'
```

```python
#checking value count for errors 

df_ords_prods['region'].value_counts(dropna = False)
```

```python
df_ords_prods[['state', 'region']].head()
```

```python
#creating crosstab for 'region' and 'spending_flag'
spending_region_crosstab = pd.crosstab(df_ords_prods['region'], df_ords_prods['spending_flag'], dropna = False)
```

```python
display(spending_region_crosstab)
```

```python
#grouping each spending flag by region for graphing purposes
grouped = df_ords_prods.groupby(['spending_flag', 'region']).size().unstack(fill_value=0)
```

```python
#creating the groups

high_spenders_count = grouped.loc['High Spender']
low_spenders_count = grouped.loc['Low Spender']
```

```python
#creating a bar chart for high spenders and saving it as an object

bar_regional_high_spenders = high_spenders_count.plot.bar(edgecolor = 'black', color = ['lightskyblue', 'peachpuff', 'palegreen', 'lavender'])
plt.title('High Spenders by Region')
plt.xlabel('Region')
plt.ylabel('Customer Count')
plt.xticks(rotation=0)
plt.show()
```

```python
#saving as a png 
bar_regional_high_spenders.figure.savefig(os.path.join(path, 'Analysis', 'Visualizations', 'bar_regional_high_spenders.png'))
```

```python
#creating a bar chart for low spenders and saving it as an object

bar_regional_low_spenders = low_spenders_count.plot.bar(edgecolor = 'black', color = ['lightskyblue', 'peachpuff', 'palegreen', 'lavender'])
plt.title('Low Spenders by Region')
plt.xlabel('Region')
plt.ylabel('Customer Count')
plt.xticks(rotation=0)
plt.show()
```

```python
#saving as a png 
bar_regional_low_spenders.figure.savefig(os.path.join(path, 'Analysis', 'Visualizations', 'bar_regional_low_spenders.png'))
```

```python
# creating a total regional customers filter 

total_regional_customers = df_ords_prods['region'].value_counts()
```

```python
#checking output 

total_regional_customers.head()
```

```python
#creating ratio for high spenders 

high_spenders_count / total_regional_customers
```

```python
#saving as a png 
bar_high_spenders_ratio.figure.savefig(os.path.join(path, 'Analysis', 'Visualizations', 'bar_high_spenders_ratio.png'))
```

```python
#converting results to percentage 

low_spenders_ratio = (low_spenders_count / total_regional_customers) * 100
```

```python
#checking ouptut 

low_spenders_ratio
```

```python
#creating a bar chart for low_spenders_ratio and saving it as an object

bar_low_spenders_ratio = low_spenders_ratio.plot.bar(edgecolor = 'black', color = ['lightskyblue', 'peachpuff', 'palegreen', 'lavender'])
plt.title('Percentage of Low Spenders by Region')
plt.xlabel('Region')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.show()
```

```python
#saving as a png 
bar_low_spenders_ratio.figure.savefig(os.path.join(path, 'Analysis', 'Visualizations', 'bar_low_spenders_ratio.png'))
```

```python
#creating an object with both high and low spenders and normalizing results 

region_spending = df_ords_prods.groupby('region')['spending_flag'].value_counts(normalize=True).unstack() * 100
```

```python
# picking colors for stacked bar chart 
colors = ['palegreen', 'lavender']
```

```python
# creating a stacked bar chart
region_spender_ratio_stacked_bar = region_spending.plot(kind='bar', stacked=True, edgecolor = 'black', color=colors)
# Add Title and Labels
region_spender_ratio_stacked_bar.set_xlabel('Region')
region_spender_ratio_stacked_bar.set_ylabel('Percent')
region_spender_ratio_stacked_bar.set_title('High and Low Spender Ratio by Region')
region_spender_ratio_stacked_bar.legend(title= 'Spending Flag', loc='upper right')
plt.xticks(rotation=0)
plt.show()
```

```python
#saving as a png 
region_spender_ratio_stacked_bar.figure.savefig(os.path.join(path, 'Analysis', 'Visualizations', 'region_spender_ratio_stacked_bar.png'))
```
